```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df =pd.read_csv('C:/newfile.csv')
df = df.set_index(['Year'])
a=pd.DataFrame(df['TOTAL'])
plt.plot(a, color='red')
plt.ylabel('Filings')
plt.xlabel('Year')
plt.title('U.S. Bankruptcy Filings Q4 2007-2016')
plt.show()
```

```python
df = df.drop(['TOTAL'], axis=1)
plt.figure(figsize=(15, 10))
plt.ylabel('Filings')
plt.xlabel('State')
df.boxplot();
```

```python
y =pd.DataFrame(df.max())
y.columns= ['Filings']
highest =pd.DataFrame(y.iloc[1:6])
print(highest)
```

```python
highest.plot(kind='bar')
plt.ylabel('Filings')
plt.xlabel('State')
plt.title('Highest Bankruptcy Filings Reported By States')
plt.show()
```

```python
x =pd.DataFrame(df.min())
x.columns= ['Filings']
lowest = pd.DataFrame(x.iloc[45:])
lowest.plot(kind='bar')
plt.ylabel('Filings')
plt.xlabel('State')
plt.title('Lowest Bankruptcy Filings Reported By States')
plt.show()
```
